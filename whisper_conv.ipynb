```python
import whisper
```

```python
model = whisper.load_model("large")
```

```python
result = model.transcribe("test.wav", task="translate", verbose=True)
```

```python
segments = result['segments']
```

```python
from pyannote.audio import Pipeline

pipeline = Pipeline.from_pretrained(
    "pyannote/speaker-diarization@2.1",
    use_auth_token="hf_"
)

diarization = pipeline("test.wav")
```

```python
speaker_turns = []

for turn, _, speaker in diarization.itertracks(yield_label=True):
    text = ""
    for seg in segments:
        if seg["start"] >= turn.start and seg["end"] <= turn.end:
            text += seg["text"].strip() + " "
    if text:
        speaker_turns.append((speaker, text.strip()))
```

```python
from deepmultilingualpunctuation import PunctuationModel

punct_model = PunctuationModel()

def punctuate(text):
    try:
        return punct_model.restore_punctuation(text)
    except:
        return text

speaker_turns_punct = [(spk, punctuate(txt)) for spk, txt in speaker_turns]
```

```python
speaker_map = {
    "SPEAKER_00": "Agent",
    "SPEAKER_01": "User"
}

formatted = ""
for speaker, text in speaker_turns_punct:
    name = speaker_map.get(speaker, speaker)
    formatted += f"{name}: {text.strip()}\n\n"

print(formatted)
```
