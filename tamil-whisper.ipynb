```python
import nltk
nltk.download('punkt') # Ensure this is downloaded
nltk.download('punkt_tab') # Download the missing resource

from nltk.tokenize import sent_tokenize
from transformers import pipeline

# Load the grammar correction model
grammar_fix = pipeline("text2text-generation", model="pszemraj/flan-t5-large-grammar-synthesis", device=0)

# Step 1: Break ASR transcript into sentences
raw_text = result
sentences = sent_tokenize(raw_text)

# Step 2: Group sentences into ~400-token chunks
chunks = []
chunk = ""
for sent in sentences:
    if len((chunk + sent).split()) < 150:  # ~150 words ≈ 400 tokens
        chunk += " " + sent
    else:
        chunks.append(chunk.strip())
        chunk = sent
if chunk:
    chunks.append(chunk.strip())

# Step 3: Run grammar fix on each chunk
corrected = []
for ch in chunks:
    output = grammar_fix(ch)[0]['generated_text']
    corrected.append(output)

# Step 4: Join and format
final_text = "\n\n".join(corrected)
```
