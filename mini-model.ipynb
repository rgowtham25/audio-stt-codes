```python
def get_audio_duration(p):
    try:
        out = subprocess.run(
            ["ffprobe","-v","error","-show_entries","format=duration","-of","default=nk=1:nw=1", p],
            capture_output=True, text=True, check=True
        )
        return float(out.stdout.strip())
    except Exception:
        return 0.0

def run_ffmpeg(cmd):
    try:
        subprocess.run(cmd, check=True, capture_output=True, text=True)
        return True
    except subprocess.CalledProcessError as e:
        print(f"FFmpeg error: {e}")
        return False

def audio_pre_v1(i, o):
    print("--- Advanced preprocessing ---")
    return run_ffmpeg([
        "ffmpeg","-y","-i",i,"-acodec","pcm_s16le","-ac","1","-ar","16000",
        "-af","loudnorm=I=-23:TP=-2:LRA=7,highpass=f=80,lowpass=f=8000,afftdn=nr=8", o
    ])

def audio_pre_v2(i, o):
    print("--- Simplified preprocessing ---")
    return run_ffmpeg([
        "ffmpeg","-y","-i",i,"-acodec","pcm_s16le","-ac","1","-ar","16000",
        "-af","loudnorm=I=-23:TP=-2,highpass=f=100", o
    ])

def audio_pre_v3(i, o):
    print("--- Basic preprocessing ---")
    return run_ffmpeg(["ffmpeg","-y","-i",i,"-acodec","pcm_s16le","-ac","1","-ar","16000","-af","loudnorm", o])

def audio_pre_v4(i, o):
    print("--- Minimal processing ---")
    return run_ffmpeg(["ffmpeg","-y","-i",i,"-acodec","pcm_s16le","-ac","1","-ar","16000", o])

def smart_pre(i, o):
    orig = get_audio_duration(i)
    print(f"Original duration: {orig:.2f}s")

    for k, fn in enumerate([audio_pre_v1, audio_pre_v2, audio_pre_v3, audio_pre_v4], start=1):
        if fn(i, o) and os.path.exists(o):
            proc = get_audio_duration(o)
            print(f"Method {k} - Processed duration: {proc:.2f}s")
            if abs(proc - orig) < 1.0:
                print(f"✅ Preprocessing successful (method {k})")
                return True
            else:
                print("⚠️ Duration mismatch, trying next method...")

    print("❌ All preprocessing methods failed")
    return False

def post_process_text(text):
    """Enhanced post-processing for insurance call context"""
    if not text:
        return ""

    # Remove generic AI responses that Voxtral tends to generate
    generic_responses = [
        "sure, i can help with that",
        "please provide me with the details you need assistance with",
        "i'm sorry, but i don't have the necessary tools",
        "however, i can guide you on how to find",
        "here are a few steps you can take",
        "if you provide more details",
        "i might be able to offer more specific guidance",
        "sure, i'd be happy to help with that",
        "however, i need a bit more context",
        "could you please provide more details",
        "i'm sorry, i didn't understand that",
        "could you please repeat it",
        "in general, when it comes to insurance policies",
        "here are a few things to keep in mind",
        "if you're referring to a specific type of policy"
    ]

    text_lower = text.lower()
    for generic in generic_responses:
        text_lower = text_lower.replace(generic, "")

    # Remove excessive repetitions
    words = text_lower.split()
    cleaned, i = [], 0
    while i < len(words):
        w = words[i]; rep = 1; j = i + 1
        while j < len(words) and words[j] == w:
            rep += 1; j += 1
        # Keep maximum 2 repetitions for normal speech patterns
        keep = min(rep, 2) if rep <= 3 else 1
        cleaned.extend(words[i:i+keep])
        i += rep

    text = ' '.join(cleaned)

    # Remove filler sounds
    filler_sounds = ["uh", "um", "mm", "hmm", "ah", "oh", "huh", "ha ha", "er", "eh"]
    for filler in filler_sounds:
        text = re.sub(rf'\b{re.escape(filler)}\b', '', text, flags=re.IGNORECASE)

    # Insurance domain corrections
    corrections = {
        'max life': 'Maxlife',
        'maxlife insurance': 'Maxlife',
        'police number': 'policy number',
        'police': 'policy',
        'some assured': 'sum assured',
        'premium do': 'premium due',
        'do date': 'due date',
        'paid up': 'paid-up',
        'payed': 'paid',
        'charges also added': 'charges are also added',
        'late fee charges': 'late fee charges',
        'health declaration form': 'health declaration form',
        'branch sir': 'branch, sir',
        'online sir': 'online, sir',
    }

    for wrong, correct in corrections.items():
        text = text.replace(wrong, correct)

    # Fix currency
    text = re.sub(r'\b(rs|rupees?)\s*', '₹', text, flags=re.IGNORECASE)

    # Clean up spaces and punctuation
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'\s+([,.!?])', r'\1', text)

    # Capitalize sentences
    text = re.sub(r'(^|[.!?]\s+)([a-z])', lambda m: m.group(1) + m.group(2).upper(), text)

    return text.strip()
```

```python
# diarization & transcription windowing
MIN_SEG_DUR = 0.6     # skip tiny segments
MAX_SEG_DUR = 20.0    # chunk any segment longer than this
CHUNK_STEP = 14.0     # inner chunk length (excl. overlap)
CHUNK_OVERLAP = 1.0   # seconds

# merge consecutive dialogue turns of the same speaker if the gap < MERGE_GAP
MERGE_GAP = 3.0

# decoding knobs (beam search, no temperature to avoid warnings)
GEN_KW = dict(
    do_sample=False,
    num_beams=4,
    early_stopping=True,
    repetition_penalty=1.25,
    no_repeat_ngram_size=4,
    max_new_tokens=320,
    # length_penalty=1.0,   # optionally tune
)
```

```python
def get_audio_duration(p):
    try:
        out = subprocess.run(
            ["ffprobe","-v","error","-show_entries","format=duration","-of","default=nk=1:nw=1", p],
            capture_output=True, text=True, check=True
        )
        return float(out.stdout.strip())
    except Exception:
        return 0.0

def run_ffmpeg(cmd):
    try:
        subprocess.run(cmd, check=True, capture_output=True, text=True)
        return True
    except subprocess.CalledProcessError:
        return False

def audio_pre_v1(i,o):
    print("--- Advanced preprocessing ---")
    return run_ffmpeg([
        "ffmpeg","-y","-i",i,"-acodec","pcm_s16le","-ac","1","-ar","16000",
        "-af","loudnorm=I=-23:TP=-2:LRA=7,highpass=f=80,lowpass=f=8000,afftdn=nr=10", o
    ])

def audio_pre_v2(i,o):
    print("--- Simplified preprocessing ---")
    return run_ffmpeg([
        "ffmpeg","-y","-i",i,"-acodec","pcm_s16le","-ac","1","-ar","16000",
        "-af","loudnorm=I=-23:TP=-2,highpass=f=100", o
    ])

def audio_pre_v3(i,o):
    print("--- Basic preprocessing ---")
    return run_ffmpeg(["ffmpeg","-y","-i",i,"-acodec","pcm_s16le","-ac","1","-ar","16000","-af","loudnorm", o])

def audio_pre_v4(i,o):
    print("--- Minimal processing ---")
    return run_ffmpeg(["ffmpeg","-y","-i",i,"-acodec","pcm_s16le","-ac","1","-ar","16000", o])

def smart_pre(i,o):
    orig = get_audio_duration(i)
    for k,fn in enumerate([audio_pre_v1,audio_pre_v2,audio_pre_v3,audio_pre_v4], start=1):
        if fn(i,o) and os.path.exists(o):
            proc = get_audio_duration(o)
            print(f"Durations: orig={orig:.2f}s processed={proc:.2f}s")
            if abs(proc-orig) < 1.0:
                print(f"✅ Preprocessing ok (method {k})")
                return True
            else:
                print("⚠️ Duration mismatch, trying next…")
    print("❌ All preprocessing methods failed")
    return False

def post_process_text(text):
    if not text:
        return ""
    # 1) collapse immediate repetitions conservatively
    words = text.split()
    cleaned, i = [], 0
    while i < len(words):
        w = words[i].lower(); rep = 1; j = i+1
        while j < len(words) and words[j].lower() == w:
            rep += 1; j += 1
        keep = min(rep, 2) if rep <= 3 else 1
        cleaned.extend(words[i:i+keep]); i += rep
    text = ' '.join(cleaned)

    # 2) remove fillers
    filler_sounds = ["uh","um","mm","hmm","ah","oh","huh","ha ha"]
    soft_fillers  = ["okay okay","yes yes","yes yes yes","i mean","you know","like like","ok ok"]
    for f in filler_sounds + soft_fillers:
        text = re.sub(rf'\b{re.escape(f)}\b','',text,flags=re.IGNORECASE)

    # 3) domain corrections
    corrections = {
        'access max life':'Axis Max Life','axis max life':'Axis Max Life',
        'g pay':'GPay','google pay':'Google Pay','phone pay':'PhonePe','phone pe':'PhonePe',
        'pay tm':'Paytm','net banking':'netbanking','some assured':'sum assured',
        'premium do':'premium due','do date':'due date','okay sir':'Okay sir',
    }
    low = text.lower()
    for wrong,correct in corrections.items():
        low = low.replace(wrong, correct)

    # Rs. -> ₹
    low = re.sub(r'\brs[.]?\s*','₹',low)

    # 4) punctuation + spaces
    low = re.sub(r'\s{2,}',' ',low)
    low = re.sub(r'[,]{2,}',',',low)
    low = re.sub(r'\s+,',',',low)
    low = re.sub(r'\s+\.','.',low)
    low = re.sub(r'\s+[!?]',lambda m:m.group(0).strip(),low)

    # 5) sentence-cap
    low = re.sub(r'(^|[.!?]\s+)([a-z])',lambda m:m.group(1)+m.group(2).upper(),low)
    return low.strip()

def force_english_chars(s: str) -> str:
    # Keep ASCII, whitespace, digits, common punct, and ₹/£/€
    return re.sub(r"[^\x00-\x7F₹£€%.,:;!?()'\"/\-\n 0-9A-Za-z]", "", s)

# remove generic assistant-y boilerplate that sometimes appears
ASSISTANTISMS = [
    "i can help with that", "i would be happy to help", "how can i assist you",
    "please provide me with the details", "here are a few steps", "consult a financial advisor",
    "i'm sorry, but i don't have", "feel free to ask", "sure, please go ahead",
]
def strip_assistantisms(s: str) -> str:
    t = s
    for p in ASSISTANTISMS:
        t = re.sub(rf'(^|[.?!]\s*)[^.?!]*{re.escape(p)}[^.?!]*', '', t, flags=re.IGNORECASE)
    # cleanup extra spaces/dots left behind
    t = re.sub(r'\s{2,}', ' ', t).strip(' .\n')
    return t

def seconds_to_srt_time(t: float) -> str:
    h = int(t // 3600)
    m = int((t % 3600) // 60)
    s = int(t % 60)
    ms = int((t - int(t)) * 1000)
    return f"{h:02d}:{m:02d}:{s:02d},{ms:03d}"
```

```python
# --- Replace your voxtral_transcribe_segment(...) with this:
import tempfile, soundfile as sf

def voxtral_chat_transcribe(audio_np: np.ndarray, sr: int) -> str:
    """Transcribe/translate to English using chat template with strict instruction."""
    with tempfile.NamedTemporaryFile(suffix=".wav", delete=True) as f:
        sf.write(f.name, audio_np, sr)

        conversation = [{
            "role": "user",
            "content": [
                {"type": "audio", "path": f.name},
                {"type": "text", "text":
                 ("Task: produce a VERBATIM transcript of the audio in ENGLISH ONLY.\n"
                  "- If speech is Tamil, translate to natural English.\n"
                  "- Domain: insurance call center (policy number, due date, premium, sum assured, late fee, payment modes).\n"
                  "- Use digits for numbers and the symbol ₹ for rupees.\n"
                  "- Output ONLY what was said. No advice, no explanations, no lists.")}
            ],
        }]

        # NOTE: use continue_final_message instead of add_generation_prompt
        inputs = processor.apply_chat_template(
            conversation,
            continue_final_message=True,
            return_tensors="pt"
        ).to(DEVICE)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            do_sample=False,
            num_beams=4,
            early_stopping=True,
            repetition_penalty=1.25,
            no_repeat_ngram_size=4,
            max_new_tokens=320,
        )

    decoded = processor.batch_decode(outputs[:, inputs.input_ids.shape[1]:], skip_special_tokens=True)
    text = decoded[0].strip() if decoded else ""
    # optional: your cleaners
    text = re.sub(r"[^\x00-\x7F₹£€%.,:;!?()'\"/\-\n 0-9A-Za-z]", "", text)  # force english-ish
    text = post_process_text(text)
    return text
```

```python
# =========================
# MAIN
# =========================
def main():
    print("🎯 Voxtral Mini + Pyannote diarized pipeline (TA→EN)")
    print("="*60)

    # 1) Preprocess
    if not smart_pre(INPUT_AUDIO_PATH, CLEAN_AUDIO_PATH):
        return

    # 2) Speaker diarization
    print("\n--- Speaker Diarization ---")
    try:
        diar = Pipeline.from_pretrained("pyannote/speaker-diarization-3.1", use_auth_token=HF_TOKEN)
        if torch.cuda.is_available():
            diar.to(torch.device("cuda")); print("✅ Using GPU for diarization")
        diarization = diar(CLEAN_AUDIO_PATH)
        print("✅ Diarization done")
    except Exception as e:
        print(f"⚠️ Diarization failed: {e}")
        diarization = None

    dialogue = []
    audio, sr = librosa.load(CLEAN_AUDIO_PATH, sr=16000, mono=True)

    if diarization is None:
        # single-shot
        raw = voxtral_chat_transcribe(audio.astype(np.float32), sr)
        if raw:
            dialogue.append({
                "speaker": "Speaker_Unknown",
                "text": raw,
                "start_time": 0.0,
                "end_time": get_audio_duration(CLEAN_AUDIO_PATH)
            })
    else:
        print("\n--- Transcribing by diarization turns (chat mode) ---")
        for turn, _, spk in diarization.itertracks(yield_label=True):
            start_s = max(0.0, float(turn.start))
            end_s   = max(start_s, float(turn.end))
            dur = end_s - start_s
            if dur < MIN_SEG_DUR:
                continue

            def transcribe_span(t0, t1):
                s_idx, e_idx = int(t0*sr), int(t1*sr)
                seg_audio = audio[s_idx:e_idx].astype(np.float32)
                if seg_audio.size == 0:
                    return None
                return voxtral_chat_transcribe(seg_audio, sr)

            # chunk long segments
            if dur > MAX_SEG_DUR:
                t = start_s
                while t < end_s:
                    t2 = min(end_s, t + CHUNK_STEP + CHUNK_OVERLAP)
                    txt = transcribe_span(t, t2)
                    if txt and len(txt) > 3:
                        dialogue.append({
                            "speaker": spk, "text": txt,
                            "start_time": float(t), "end_time": float(t2)
                        })
                    t += CHUNK_STEP
            else:
                txt = transcribe_span(start_s, end_s)
                if txt and len(txt) > 3:
                    dialogue.append({
                        "speaker": spk, "text": txt,
                        "start_time": start_s, "end_time": end_s
                    })

        # 3) merge consecutive same-speaker turns
        merged = []
        for d in dialogue:
            if merged and merged[-1]["speaker"] == d["speaker"] and d["start_time"] - merged[-1]["end_time"] < MERGE_GAP:
                merged[-1]["text"] += " " + d["text"]
                merged[-1]["end_time"] = d["end_time"]
            else:
                merged.append(d)
        dialogue = merged

    # 4) pretty print
    print("\n🎭 DIALOGUE OUTPUT " + "="*40)
    for d in dialogue:
        print(f"\n{d['speaker']} [{d['start_time']:.1f}s - {d['end_time']:.1f}s]:")
        print("  📝", d["text"])

    # 5) save JSON
    out = {
        "metadata": {
            "model_used": MODEL_ID,
            "sample_rate": 16000,
            "segments": len(dialogue),
            "total_duration": get_audio_duration(CLEAN_AUDIO_PATH),
            "processing_successful": True,
            "mode": "chat_instruct_TA_to_EN"
        },
        "dialogue": dialogue
    }
    with open("voxtral_diarized_results.json","w",encoding="utf-8") as f:
        json.dump(out, f, indent=2, ensure_ascii=False)
    print("\n💾 Saved: voxtral_diarized_results.json")

    # 6) save SRT
    with open("voxtral_diarized_results.srt","w",encoding="utf-8") as f:
        for idx, d in enumerate(dialogue, start=1):
            start = seconds_to_srt_time(d["start_time"])
            end = seconds_to_srt_time(d["end_time"])
            f.write(f"{idx}\n{start} --> {end}\n{d['speaker']}: {d['text']}\n\n")
    print("💾 Saved: voxtral_diarized_results.srt")

if __name__ == "__main__":
    main()
```
