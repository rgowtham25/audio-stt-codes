```python
import os
import torch
from faster_whisper import WhisperModel
from deepmultilingualpunctuation import PunctuationModel
from datetime import timedelta
```

```python
input_audio_path = "test.wav"
clean_audio_path = "cleaned_audio.wav"
```

```python
os.system(
    f'ffmpeg -y -i "{input_audio_path}" '
    f'-acodec pcm_s16le -ac 1 -ar 16000 '
    f'-af "loudnorm,highpass=f=200,lowpass=f=3000" '
    f'"{clean_audio_path}"'
)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model = WhisperModel("large-v3", device=device, compute_type="int8" if device == "cpu" else "float16")
```

```python
segments, info = model.transcribe(
    clean_audio_path,
    task="translate",       # Ensures output is English
    beam_size=5,
    best_of=5,
    word_timestamps=True
)
```

```python
transcript = []
speaker_id = 0

for seg in segments:
    speaker = f"Sp{speaker_id % 2 + 1}"
    start = str(timedelta(seconds=int(seg.start)))
    end = str(timedelta(seconds=int(seg.end)))
    line = f"{speaker} [{start} - {end}]: {seg.text.strip()}"
    transcript.append(line)
    speaker_id += 1
```

```python
punct_model = PunctuationModel()
combined_text = " ".join([line.split(": ", 1)[-1] for line in transcript])
punctuated = punct_model.restore_punctuation(combined_text).split(". ")

# Map back to transcript structure
final_transcript = []
for i, line in enumerate(transcript):
    prefix = line.split(":")[0]
    sentence = punctuated[i].strip() if i < len(punctuated) else ""
    final_transcript.append(f"{prefix}: {sentence}.")
```

```python
with open("final_transcript.txt", "w", encoding="utf-8") as f:
    for line in final_transcript:
        f.write(line + "\n")

print("✅ Transcription complete. Output saved to final_transcript.txt")
```
