```python
result_transcribe = model.transcribe("001_t1.wav", task="transcribe")
```

```python
native_text = result_transcribe["text"]
```

```python
# Connect to OpenRouter using OpenAI SDK
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key="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"
)

# Send to Sarvam-M for translation to English dialogue
response = client.chat.completions.create(
    model="sarvamai/sarvam-m:free",
    messages=[
        {"role": "system", "content": "You are a translator that converts Indian regional languages to English. Output the result as a conversation with clear speaker turns like '[Speaker 1]: ...'"},
        {"role": "user", "content": native_text}
    ]
)

# Print the English output
print("🗣️ Translated Dialog:\n")
print(response.choices[0].message.content)
```

```python
translated_dialog = response.choices[0].message.content
print("🗣️ Translated Dialog:\n")
print(translated_dialog)
```
