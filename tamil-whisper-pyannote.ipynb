```python
import whisper
from pyannote.audio import Pipeline
import torch
from pydub import AudioSegment
import re
import os # For managing files
import subprocess # Use subprocess for better error handling than os.system
```

```python
input_audio_path = "/content/001_t2.wav"
# Cleaned audio will be used by both Whisper and pyannote.audio
clean_audio_path = "cleaned_audio_for_asr_and_diarization.wav"
```

```python
# --- Step 1: Enhanced Audio Preprocessing with ffmpeg ---
print("--- Starting Audio Preprocessing ---")
ffmpeg_command = [
    "ffmpeg",
    "-i", input_audio_path,
    "-acodec", "pcm_s16le",
    "-ac", "1",
    "-ar", "16000",
    # Removed 't=1000' from anlmdn
    "-af", "loudnorm=I=-16:TP=-1.5:LRA=11, highpass=f=200, lowpass=f=3000, anlmdn=s=0.5",
    "-y", clean_audio_path
]

try:
    result = subprocess.run(ffmpeg_command, check=True, capture_output=True, text=True)
    print(result.stdout)
    if result.stderr:
        print("FFmpeg stderr (might contain warnings/info, not necessarily errors):")
        print(result.stderr)
    print(f"--- Audio Preprocessing Complete. Cleaned audio saved to {clean_audio_path} ---")
except subprocess.CalledProcessError as e:
    print(f"--- FFmpeg Error ---")
    print(f"Command: {' '.join(e.cmd)}")
    print(f"Return Code: {e.returncode}")
    print(f"STDOUT:\n{e.stdout}")
    print(f"STDERR:\n{e.stderr}")
    print(f"--- Audio Preprocessing Failed. Cannot proceed. ---")
    raise e
except FileNotFoundError:
    print("--- FFmpeg not found ---")
    print("Please ensure FFmpeg is installed and accessible in your environment's PATH.")
    raise
```

```python
print("--- Whisper Transcription Complete ---")
```

```python
print("--- Starting Speaker Diarization ---")
```

```python
pipeline = Pipeline.from_pretrained(
    "pyannote/speaker-diarization-3.1",
    use_auth_token=HUGGING_FACE_ACCESS_TOKEN
)

if torch.cuda.is_available():
    pipeline.to(torch.device("cuda"))
    print("Pyannote.audio moved to GPU.")
else:
    print("CUDA not available, running pyannote.audio on CPU. This might be slow.")

diarization = pipeline(clean_audio_path)
print("--- Speaker Diarization Complete ---")
```

```python
# Helper function to find the speaker for a given time segment
def get_speaker_for_segment(start_time, end_time, diarization_result):
    speakers = {}
    for segment, _, speaker_label in diarization_result.itertracks(yield_label=True):
        # Calculate overlap
        overlap_start = max(start_time, segment.start)
        overlap_end = min(end_time, segment.end)
        overlap_duration = max(0, overlap_end - overlap_start)

        if overlap_duration > 0:
            speakers[speaker_label] = speakers.get(speaker_label, 0) + overlap_duration

    if not speakers:
        return "Unknown Speaker"

    # Return the speaker with the most overlap
    return max(speakers, key=speakers.get)

print("\n--- Generating Dialogue Output ---")
dialogue_output = []
for segment in whisper_result["segments"]:
    start = segment['start']
    end = segment['end']
    text = segment['text'].strip()

    # Assign speaker based on the segment's time range
    speaker = get_speaker_for_segment(start, end, diarization)
    dialogue_output.append(f"Speaker {speaker}: {text}")
```

```python
# --- Step 5: Post-processing (Basic) ---
import re
final_dialogue = []
previous_speaker = None
for line in dialogue_output:
    current_speaker_match = re.match(r"Speaker (SPEAKER_\d+):", line)
    if current_speaker_match:
        current_speaker = current_speaker_match.group(1)
        current_text = line.split(':', 1)[1].strip()

        # Simple check for immediate, exact repetitions (can be improved)
        if final_dialogue and previous_speaker == current_speaker:
            # Get text of the last entry by the same speaker
            last_entry_text = final_dialogue[-1].split(':', 1)[1].strip()
            if last_entry_text == current_text:
                # If exact repetition, skip adding this line
                continue

        final_dialogue.append(line)
        previous_speaker = current_speaker
    else:
        # If line doesn't match expected format, just add it
        final_dialogue.append(line)


# Print the final dialogue
for line in final_dialogue:
    print(line)

print("\n--- Processing Complete ---")
```

```python
print("--- Starting Audio Preprocessing ---")
ffmpeg_command = [
    "ffmpeg",
    "-i", input_audio_path,
    "-acodec", "pcm_s16le",
    "-ac", "1",
    "-ar", "16000",
    # Reverting to the original working filters: loudnorm, highpass, lowpass
    "-af", "loudnorm=I=-16:TP=-1.5:LRA=11, highpass=f=200, lowpass=f=3000",
    "-y", clean_audio_path
]

try:
    # Use subprocess.run for better error handling and blocking execution
    result = subprocess.run(ffmpeg_command, check=True, capture_output=True, text=True)
    # Print FFmpeg's stdout (if any)
    if result.stdout:
        print("FFmpeg stdout:")
        print(result.stdout)
    # Print FFmpeg's stderr (often contains warnings/info, not necessarily errors)
    if result.stderr:
        print("FFmpeg stderr (might contain warnings/info):")
        print(result.stderr)
    print(f"--- Audio Preprocessing Complete. Cleaned audio saved to {clean_audio_path} ---")

    # Optional: Verify duration of the cleaned audio file using ffprobe
    print(f"--- Verifying Cleaned Audio Duration ---")
    ffprobe_command = ["ffprobe", "-v", "error", "-show_entries", "format=duration", "-of", "default=noprint_wrappers=1:nokey=1", clean_audio_path]
    duration_output = subprocess.run(ffprobe_command, check=True, capture_output=True, text=True).stdout.strip()
    try:
        duration_seconds = float(duration_output)
        print(f"Cleaned audio duration: {duration_seconds:.2f} seconds")
    except ValueError:
        print(f"Could not parse duration from ffprobe: {duration_output}")

except subprocess.CalledProcessError as e:
    print(f"--- FFmpeg Error ---")
    print(f"Command: {' '.join(e.cmd)}")
    print(f"Return Code: {e.returncode}")
    print(f"STDOUT:\n{e.stdout}")
    print(f"STDERR:\n{e.stderr}")
    print(f"--- Audio Preprocessing Failed. Cannot proceed. ---")
    raise e
except FileNotFoundError:
    print("--- FFmpeg/FFprobe not found ---")
    print("Please ensure FFmpeg and FFprobe are installed and accessible in your environment's PATH.")
    raise
```
