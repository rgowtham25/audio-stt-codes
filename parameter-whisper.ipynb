```python
import whisper
```

```python
model = whisper.load_model("large")
```

```python
result = model.transcribe("test.wav",
    task="translate",            # Translates Telugu → English
    temperature=0.5,             # Mid-range for noisy speech
    best_of=5,                   # Sample multiple candidates, pick best
    beam_size=5,                 # For beam search decoding
    fp16=False,                  # Force FP32 for Colab CPU compatibility
    verbose=True,
    word_timestamps=True        # Optional, helps in diarization logic
)
```

```python
result["text"]
```

```python
def format_as_dialogue(segments):
    dialogue = []
    speakers = ["Agent", "User"]
    current = 0

    for seg in segments:
        text = seg["text"].strip()
        if text:
            dialogue.append(f"{speakers[current]}: {text}")
            current = 1 - current  # Alternate speakers

    return "\n".join(dialogue)
```

```python
print(format_as_dialogue(result["segments"]))
```
